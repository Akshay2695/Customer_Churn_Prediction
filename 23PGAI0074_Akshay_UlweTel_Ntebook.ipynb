# Logistic_Regression

```python
df1 = df1.replace(r'^\s*$', np.nan, regex=True)
```

```python
# We will use the data frame where we had created dummy variables
y = df1['churn'].values
X = df1.drop(columns = ['churn'])

# Scaling all the variables to a range of 0 to 1
from sklearn.preprocessing import MinMaxScaler
features = X.columns.values
scaler = MinMaxScaler(feature_range = (0,1))
scaler.fit(X)
X = pd.DataFrame(scaler.transform(X))
X.columns = features
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
result = model.fit(X_train, y_train)
```

```python
from sklearn import metrics
prediction_test = model.predict(X_test)
# Print the prediction accuracy
print (metrics.accuracy_score(y_test, prediction_test))
```

```python
weights = pd.Series(model.coef_[0],
                 index=X.columns.values)
print (weights.sort_values(ascending = False)[:10].plot(kind='bar'))
```

```python
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")

# Creating the hyperparameter grid
param_grid = {'C': [0.1, 1, 10, 100], 'penalty': ['l1', 'l2']}

# Initializing the logistic regression model
model = LogisticRegression()

# Initializing the GridSearchCV object
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5)

# Fitting the GridSearchCV object to the training data
grid_search.fit(X_train, y_train)

# Printing the best hyperparameters
print("Best hyperparameters: ", grid_search.best_params_)
```

```python
# Using the best model to predict the labels of the test data
best_model = grid_search.best_estimator_
test_predictions = best_model.predict(X_test)

# Calculating the accuracy of the best model on the test data
from sklearn import metrics
accuracy = metrics.accuracy_score(y_test, test_predictions)
print("Accuracy of the best model on the test data: ", accuracy)
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics

best_model = grid_search.best_estimator_
y_pred_proba = best_model.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="AUC="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

best_model = grid_search.best_estimator_
y_pred_proba = best_model.predict_proba(X_test)[::,1]
precision, recall, thresholds = precision_recall_curve(y_test, y_pred_proba)

plt.plot(recall, precision, 'g-', label='AUC = {:.2f}'.format(metrics.auc(recall, precision)))
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend()
plt.show()
```

# Logistic_regression with PCA

```python
from sklearn.decomposition import PCA

# Create a PCA instance with the desired number of components
pca = PCA(n_components=25)

# Fit and transform the data
X_pca = pca.fit_transform(X)
```

```python
print(pca.explained_variance_ratio_)
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot of the first and second principal components
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Fit PCA to the data
pca = PCA()
pca.fit(X)

# Get the explained variance ratio of each component
var_exp = pca.explained_variance_ratio_

# Create an elbow plot
plt.figure()
plt.plot(np.cumsum(var_exp))
plt.xlabel('Number of Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Elbow Plot')
plt.show()
```

```python
# Get the matrix of principal component vectors
components = pca.components_

# Create a Series from the weights of the first principal component
pc_weights = pd.Series(components[0], index=X.columns.values)

# Sort the weights and select the top 10
pc_weights = pc_weights.sort_values(ascending = False)[:10]

# Plot the weights
pc_weights.plot(kind='bar')
plt.xlabel('Features')
plt.ylabel('Principal Component Weights')
plt.title('First Principal Component')
plt.show()
```

```python
# Splitting the data into training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.3, random_state=0)

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
result = model.fit(X_train, y_train)
```

```python
prediction_test = model.predict(X_test)
# Print the prediction accuracy
print (metrics.accuracy_score(y_test, prediction_test))
```

```python
from sklearn.metrics import classification_report, confusion_matrix  
print(confusion_matrix(y_test,prediction_test)) 
```
